```python
import pandas as pd
import matplotlib.pyplot as plt

csv_path = "./bsim45.csv"

df = pd.read_csv(csv_path, delim_whitespace=True)

if 'v-sweep' in df.columns and '@m1[id]' in df.columns and 'gmid' in df.columns:
    df['v-sweep'] = df['v-sweep'].astype(float)
    df['@m1[id]'] = df['@m1[id]'].astype(float)
    df['gmid'] = df['gmid'].astype(float)
    
    fig, ax1 = plt.subplots(figsize=(10, 6))
        
    ax1.plot(df['v-sweep'], df['gmid'], label='gmid', color='blue', marker='x')
    ax1.set_xlabel('VGB [V]')
    ax1.set_ylabel('gmid', color='blue')
    ax1.tick_params(axis='y', colors='blue')
    
    ax2 = ax1.twinx()
    ax2.plot(df['v-sweep'], df['@m1[id]'], label='@m1[id]', color='red', marker='o')
    ax2.set_ylabel('@m1[id]', color='red')
    ax2.set_yscale('log')
    ax2.tick_params(axis='y', colors='red')

    fig.tight_layout()
    ax1.grid(True)

    ax1.legend(loc='upper left')
    ax2.legend(loc='upper right')
    
    plt.show()
```
